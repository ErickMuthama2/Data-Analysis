```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Loading the data
df=pd.read_excel('C:/Users/Erick/Desktop/Coffee/Coffee Shop Sales.xlsx')
```

```python
#Checking the data 
df.head()
```

```python
#checking any duplicates in the dataset.
df.duplicated().sum()
```

```python
# checking missing values in the dataset.
df.isnull().sum()
```

```python
df.info()
```

```python
df['transaction_time']=df['transaction_time'].astype(str)
```

```python
df['transaction_time']=pd.to_timedelta(df['transaction_time'])
```

```python
df.info()
```

```python
# Creating new columns for sales and datetime
df['sales']=df['transaction_qty']*df['unit_price']
df['datetime']=df['transaction_date']+df['transaction_time']
```

```python
df.head()
```

```python
#Summary Statistics
df.describe()
```

```python
sales_by_location=df.groupby('store_location').agg({
    'sales':'sum',
    'transaction_id':'count'
})

sales_by_location
```

```python
daily_transaction_by_location=df.groupby(['transaction_date','store_location'])['sales'].sum().unstack()
daily_transaction_by_location
```

```python
daily_transaction_by_location.plot(figsize=(14,8),title='Daily transaction by location')

plt.xlabel('Date')
plt.ylabel('Total sales ($)')
plt.legend(title='Store Location')
plt.grid(True)

plt.show()
                                   
```

```python
#Plotting a stackplot for daily transaction by location
plt.figure(figsize=(14,8))

plt.stackplot(daily_transaction_by_location.index,daily_transaction_by_location.T,labels=daily_transaction_by_location.columns)

plt.title('Daily Sales by store location (Stacked Area Chart)')
plt.xlabel('Date')
plt.ylabel('Total sales ($)')
plt.legend(title='Store location',loc='upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
daily_sales=df.groupby('transaction_date')['sales'].sum().reset_index()
daily_sales
```

```python
daily_sales=df.groupby('transaction_date')['sales'].sum().reset_index()
df['day_of_week']=df['datetime'].dt.day_name()

weekly_sales=df.groupby('day_of_week')['sales'].sum().reindex([
    'Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']).reset_index()

df['Month']=df['datetime'].dt.to_period('M')
monthly_sales=df.groupby('Month')['sales'].sum().reset_index()

daily_sales, weekly_sales, monthly_sales
```

```python
monthly_sales.plot(figsize=(14,8),title='Monthly Sales')

plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
weekly_sales=df.groupby(['day_of_week','store_location'])['sales'].sum().unstack()
days_order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekly_sales=weekly_sales.reindex(days_order)
```

```python
weekly_sales[["Astoria","Hell's Kitchen","Lower Manhattan"]].plot(kind='bar',figsize=(14,8))

plt.xlabel('Day of the week', fontweight='bold')
plt.ylabel('Total Sales ($)',fontweight='bold')
plt.title('Weekly Sales by store location')
plt.legend(title='Store Location',loc='upper left')
plt.tight_layout()

plt.show()
```

```python
df_category=df.groupby('product_category').agg({
    'sales':'sum',
    'transaction_id':'count'
}).sort_values('sales',ascending=False)

df_category
```

```python
df['hour']=df['datetime'].dt.hour
hourly_sales_by_category=df.groupby(['hour','product_category'])['sales'].sum().unstack()

hourly_sales_by_category.plot(kind='bar',stacked=True,figsize=(14,8),colormap='viridis')

plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales ($)')
plt.title('Hourly sales by product category')
plt.grid(True)
plt.xticks(range(24))
plt.tight_layout()

plt.show()
```

```python
df_category['avg_sales']=df_category['sales']/df_category['transaction_id']
df_category.sort_values('avg_sales',ascending=False)
```

```python
sale_by_product=df.groupby('product_type').agg({
    'sales':'sum',
    'unit_price':'mean'
}).sort_values('sales',ascending=False)

sale_by_product
```

```python
product_sales_df=df.groupby(['product_category','product_type'])['sales'].sum().reset_index()
product_sales_df
```

```python
plt.figure(figsize=(14,8))
sns.barplot(data=product_sales_df,x='product_type',y='sales',hue='product_category',palette='viridis')

plt.xlabel('Product Type')
plt.ylabel('Total Sales ($)')
plt.title('Total Sales by product Type')
plt.legend(title='Product Category',loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
categories=product_sales_df['product_category'].unique()

for category in categories:
    plt.figure(figsize=(5,5))
    category_data=product_sales_df[product_sales_df['product_category']==category]
    sns.barplot(data=category_data,x='product_type',y='sales')

    #Add labels and title
    plt.xlabel('Product Type')
    plt.ylabel('Total Sales ($)')
    plt.title(f'Total Sales by Product Type in {category}')
    plt.tight_layout()
    plt.xticks(rotation=45)
    plt.show()
```
